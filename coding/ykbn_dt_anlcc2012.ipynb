```python
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
%matplotlib inline
```

# LOADing The  DATASET

```python
data2012=pd.read_csv('../Data/IPD-Under-5-years-2012.csv')
data2013=pd.read_csv('../Data/IPD-Under-5-years-2013.csv')
data2014=pd.read_excel('../Data/IPD-Under-5-years-2014.xls')
data2015=pd.read_csv('../Data/IPD-under-5-years-2015-Council.csv')
```

# Check for Missing value

# FIRST TEN DISEASES WITH THE LARGEST INCIDENCES

```python
#The first slicing select first ten diseases with largest incidence ,and the second slicing ignore the valuable 'Total'
dt12=data2012.mean().sort_values(ascending=False)[0:11][1:]
dt13=data2013.mean().sort_values(ascending=False)[0:11][1:]
dt14=data2014.mean().sort_values(ascending=False)[0:11][1:]
dt15=data2015.mean().sort_values(ascending=False)[0:11][1:]
```

```python

fig1=plt.figure(figsize=(20,10))#define the plotting area
plt.subplot(141)
dt12.plot(kind='bar',color='k')
plt.title("top ten diseases incidence 2012")
plt.ylabel("QUANTIFICATION MEASURES")
plt.subplot(142)
dt12.plot(kind='bar',color='grey')
plt.title("top ten diseases incidence 2013")
plt.ylabel("")
plt.subplot(143)
dt12.plot(kind='bar',color='b')
plt.title("top ten diseases incidence 2014")
plt.ylabel("")
plt.subplot(144)
dt12.plot(kind='bar',color='c')
plt.title("top ten diseases incidence 2015")
plt.ylabel("")
plt.subplots_adjust(left=.2,wspace=.3,top=.5)

plt.savefig("../plots/plOTASK/toP10_diseases.png")
plt.savefig("../plots/plOTASK/toP10_diseases.pdf")
```

# VISUALIZING TOP TEN REGIONS/COUNCILS WITH HIGHEST INCIDENCE

```python
d12=data2012.groupby('Region')[['Total']].mean().sort_values(by=['Total'], ascending=False)[0:11][1:]
d13=data2013.groupby('Region')[['Total']].mean().sort_values(by=['Total'], ascending=False)[0:11][1:]
d14=data2014.groupby('Region')[['Total Diagnoses']].mean().sort_values(by=['Total Diagnoses'], ascending=False)[0:11][1:]
d15=data2015.groupby('Council')[['Total']].mean().sort_values(by=['Total'], ascending=False)[0:11][1:]
```
