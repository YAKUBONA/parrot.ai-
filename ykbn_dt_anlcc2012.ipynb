
# VISUALIZING DISEASE INCIDENCE 

```python
data2012.drop('Total',axis=1,inplace=True)
```

```python
fig=plt.figure(figsize=(18,10))#define the plotting area
data2012.mean().plot(kind='bar')
plt.title("DISEASES PREDICTION UNDER VISUALIZATION")
plt.ylabel("QUANTIFICATION MEASURES")
plt.savefig("../figure/mean_dist2012.png")
plt.savefig("../figure/mean_dist2012.pdf")
```

```python
data2013.drop('Total',axis=1,inplace=True)
```

```python
fig=plt.figure(figsize=(18,10))#define the plotting area
data2013.mean().plot(kind='bar')
plt.title("DISEASES PREDICTION UNDER VISUALIZATION")
plt.ylabel("QUANTIFICATION MEASURES")
plt.savefig("../figure/mean_dist2013.png")
plt.savefig("../figure/mean_dist2013.pdf")
```
